```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
def plot_beta(a, b):
    x = np.arange (0.01, 1, 0.01)
    y = beta.pdf(x,a,b)
    plt.plot(x,y)
    plt.xlim(0,1)
    plt.ylim(0,2.5)
plot_beta(0.5,0.5)
plot_beta(5,1)
plot_beta(1,3)
plot_beta(2,2)
plot_beta(2,5)
plt.show()
```
