```python
# Importando módulos necesarios 
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats 
import seaborn as sns
```

```python
# Ejercicio 1
# Graficando Beta
a, b = 0.5, 0.5 # parametros de forma
beta = stats.beta(a,b)
x = np.linspace(beta.ppf(0.01),
                beta.ppf(0.99), 100)
fp = beta.pdf(x) # Función de Probabilidad
plt.plot(x, fp)

a, b = 5, 1 # parametros de forma
beta = stats.beta(a,b)
x = np.linspace(beta.ppf(0.01),
                beta.ppf(0.99), 100)
fp = beta.pdf(x) # Función de Probabilidad
plt.plot(x, fp)

a, b = 1, 3 # parametros de forma
beta = stats.beta(a,b)
x = np.linspace(beta.ppf(0.01),
                beta.ppf(0.99), 100)
fp = beta.pdf(x) # Función de Probabilidad
plt.plot(x, fp)

a, b = 2, 2 # parametros de forma
beta = stats.beta(a,b)
x = np.linspace(beta.ppf(0.01),
                beta.ppf(0.99), 100)
fp = beta.pdf(x) # Función de Probabilidad
plt.plot(x, fp)

a, b = 2, 5 # parametros de forma
beta = stats.beta(a,b)
x = np.linspace(beta.ppf(0.01),
                beta.ppf(0.99), 100)
fp = beta.pdf(x) # Función de Probabilidad
plt.plot(x, fp)


plt.title('Distribución Beta')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```
